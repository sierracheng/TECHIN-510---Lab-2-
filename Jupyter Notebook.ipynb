```python
import requests
import time
```

```python
seattle_event_page_url = "https://visitseattle.org/events"
seattle_event_page_response = requests.get(seattle_event_page_url)
open("seattleevents.html", "w").write(seattle_event_page_response.text)
from bs4 import BeautifulSoup
seattle_event_page_strcut_data = BeautifulSoup(seattle_event_page_response.text, "html.parser")
seattle_event_page_strcut_data
```

```python
html_selector = "div.search-result-preview > div > h3 > a"
seattle_event_href_list = seattle_event_page_strcut_data.select(html_selector)
seattle_event_href_list
```

```python
# this is only one page of list 1
seattle_event_href_list_ans = [x['href'] for x in seattle_event_href_list]
```

```python
from bs4 import BeautifulSoup
seattle_event_href_list_ans = []
for page in range(1, 49):
    seattle_event_page_url = "https://visitseattle.org/events/page/" + str(page)
    print(seattle_event_page_url)
    seattle_event_page_response = requests.get(seattle_event_page_url)
    open("seattleevents.html", "w").write(seattle_event_page_response.text)
    seattle_event_page_strcut_data = BeautifulSoup(seattle_event_page_response.text, "html.parser")
    html_selector = "div.search-result-preview > div > h3 > a"
    seattle_event_href_list = seattle_event_page_strcut_data.select(html_selector)
    seattle_event_href_list_ans += [x['href'] for x in seattle_event_href_list]

seattle_event_href_list_ans
```
